```python
###Step 3 - Upload feature table.

uploaded2 = files.upload()
filename2 = next(iter(uploaded2))
feature_table = pd.read_csv(filename2, delimiter = ",")

```

```python
### Transform Data

def get_feature(pos):  #lookup feature from feature table.
    z=feature_table.loc[feature_table["Start"]<=pos+1].iloc[-1:]["Feature"]
    return z.item()
def max_c(input):  #max of 0 or input
    return np.max(0,input)
def process_data(df, input_seq, feature_table):  #inputs pmf.csv and input_sequence, feature table
    assert len(input_sequence)<len(df)
    s=pd.Series(list(input_sequence))
    s.name="ref_seq"
    df=pd.concat([df, s], axis=1)
    ##


    df["Feature"]=df.apply(lambda x: get_feature(x.name),axis=1)
    df["zero"]=0
    df["#_total"]=df['sum'] = df[['A','C','G','T','N','INS','DEL','SKIP']].sum(axis=1)
    df["#_match"]=df['sum'] = df[['A','C','G','T']].sum(axis=1)
    df["#_err_all"]=df['sum'] = df['#_total']-df['#_match']
    df["#_mut"]=df['sum'] = df['#_err_all']-df['N']-df['INS']-df['DEL']-df['SKIP']

    df["rel_match"]=df["#_match"]/df["#_total"]*100
    df["rel_error_all"]=df["#_err_all"]/df["#_total"]-df['ERR.1']*100
    df["rel_error_all"]=df[['zero','rel_error_all']].max(axis=1)

    df["rel_error_all"]=df["#_err_all"]/df["#_total"]-df['ERR.1']*100
    df["rel_error_all"]=df[['zero','rel_error_all']].max(axis=1)

    df["rel_mut"]=df["#_mut"]/df["#_total"]-df['ERR.1']*100
    df["rel_mut"]=df[['zero','rel_mut']].max(axis=1)

    df["rel_N"]=df["N"]/df["#_total"]-df['ERR.1']*100
    df["rel_N"]=df[['zero','rel_N']].max(axis=1)

    df["rel_INS"]=df["INS"]/df["#_total"]-df['ERR.1']*100
    df["rel_INS"]=df[['zero','rel_INS']].max(axis=1)

    df["rel_DEL"]=df["DEL"]/df["#_total"]-df['ERR.1']*100
    df["rel_DEL"]=df[['zero','rel_DEL']].max(axis=1)

    df["rel_SKIP"]=df["SKIP"]/df["#_total"]-df['ERR.1']*100
    df["rel_SKIP"]=df[['zero','rel_SKIP']].max(axis=1)
    return df

df_output=process_data(df,input_sequence,feature_table)
df_output
```

```python
###Plot Error Rates
###This is a stacked bar chart using percentage as y axis and base pair as x axis. Relative matches + mismathces should total to 1.0

fig = px.Figure()
labels=['rel_match','rel_mut','rel_N','rel_INS','rel_DEL','rel_SKIP']
for c in labels:
    fig.add_trace(go.Scatter(
    name = c,
    x=df_output["bp"],
    y=df_output[c],
    stackgroup='one'
   ))
fig.update_layout(hovermode="x",
    title="relative matches and errors",
    xaxis_title="bp",
    yaxis_title="%",)
fig.show()
```

```python
###Plot Match and Error Totals
###This is a stacked bar chart using match/error totals as y axis and base pair as x axis. #_total is NOT stacked and should match the total match/errors.

fig = px.Figure()
labels=['#_match','N','DEL','INS','SKIP']
for c in labels:
    fig.add_trace(go.Scatter(
    name = c,
    x=df_output["bp"],
    y=df_output[c],
    stackgroup='one'
   ))
fig.add_trace(go.Scatter(
    name='#_total',
    x=df_output["bp"],
    y=df_output['#_total'],
   ))
fig.update_traces(line=dict(width=0.5))
fig.update_layout(hovermode="x",title="Total Matches and Errors",
    xaxis_title="bp",
    yaxis_title="Num",)
fig.show()
```
