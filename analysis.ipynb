```python
##
def get_feature(pos):  #lookup feature from feature table.
    z=feature_table.loc[feature_table["Start"]<=pos+1].iloc[-1:]["Feature"]
    return z.item()
def max_c(input):  #max of 0 or input
    return np.max(0,input)
def process_data(df, input_seq, feature_table):  #inputs pmf.csv and input_sequence, feature table
    assert len(input_sequence)<len(df)
    s=pd.Series(list(input_sequence))
    s.name="ref_seq"
    df=pd.concat([df, s], axis=1)
    ##


    df["Feature"]=df.apply(lambda x: get_feature(x.name),axis=1)
    df["zero"]=0
    df["#_total"]=df['sum'] = df[['A','C','G','T','N','INS','DEL','SKIP']].sum(axis=1)
    df["#_match"]=df['sum'] = df[['A','C','G','T']].sum(axis=1)
    df["#_err_all"]=df['sum'] = df['#_total']-df['#_match']
    df["#_mut"]=df['sum'] = df['#_err_all']-df['N']-df['INS']-df['DEL']-df['SKIP']

    df["rel_match"]=df["#_match"]/df["#_total"]
    df["rel_error_all"]=df["#_err_all"]/df["#_total"]-df['ERR.1']
    df["rel_error_all"]=df[['zero','rel_error_all']].max(axis=1)

    df["rel_error_all"]=df["#_err_all"]/df["#_total"]-df['ERR.1']
    df["rel_error_all"]=df[['zero','rel_error_all']].max(axis=1)
    # df["rel_mut"]=np.max(0,df["#_mut"]/df["#_total"]-df['ERR.1'])
    # df["rel_N"]=np.max(0,df["N"]/df["#_total"]-df['ERR.1'])
    # df["rel_INS"]=np.max(0,df["INS"]/df["#_total"]-df['ERR.1'])
    # df["rel_DEL"]=np.max(0,df["DEL"]/df["#_total"]-df['ERR.1'])
    # df["rel_SKIP"]=np.max(0,df["SKIP"]/df["#_total"]-df['ERR.1'])
    return df

process_data(df1,input_sequence,feature_table)
```
